## Step 3 ŌĆö Review: The Original 3-Block CNN

Below is the original `PyTorchCNN` from the tutorial. **Read it carefully** and trace the shape of the data through each block before attempting the task.

### Shape Trace ŌĆö Fill In the Table

Complete the table by computing the output shape after each operation. Use the formula:

$$\text{Output size} = \left\lfloor \frac{\text{Input size} - \text{kernel size}}{\text{stride}} \right\rfloor + 1$$

For `MaxPool2d(kernel_size=2)`, output size = input size ├Ę 2.

| Layer | Operation | Output Shape (C, H, W) |
|---|---|---|
| Input | ŌĆö | (1, 28, 28) |
| Block 1 | Conv2d(1ŌåÆ3, k=5) | (3, **?**, **?**) |
| Block 1 | MaxPool2d(2) | (3, **?**, **?**) |
| Block 2 | Conv2d(3ŌåÆ16, k=3) | (16, **?**, **?**) |
| Block 2 | MaxPool2d(2) | (16, **?**, **?**) |
| Block 3 | Conv2d(16ŌåÆ32, k=3) | (32, **?**, **?**) |
| Block 3 | MaxPool2d(2) | (32, **?**, **?**) |
| Flatten | ŌĆö | (**?**,) |

> ­¤ÆĪ **Hint:** The final flattened size tells you what number to use in `Linear(?, 20)` inside `fc_layers`.

````python
# ŌöĆŌöĆ Original 3-Block CNN (from Tutorial) ŌĆö DO NOT MODIFY THIS CELL ŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆ

class PyTorchCNN(torch.nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.cnn_layers = torch.nn.Sequential(

            # Block 1 ŌĆö Input: (1, 28, 28) ŌåÆ Output: (3, 12, 12)
            torch.nn.Conv2d(1, 3, kernel_size=5),
            torch.nn.BatchNorm2d(3),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2),

            # Block 2 ŌĆö Input: (3, 12, 12) ŌåÆ Output: (16, 5, 5)
            torch.nn.Conv2d(3, 16, kernel_size=3),
            torch.nn.BatchNorm2d(16),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2),

            # Block 3 ŌĆö Input: (16, 5, 5) ŌåÆ Output: (32, 1, 1)
            torch.nn.Conv2d(16, 32, kernel_size=3),
            torch.nn.BatchNorm2d(32),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2),
        )

        self.fc_layers = torch.nn.Sequential(
            torch.nn.Linear(32, 20),
            torch.nn.BatchNorm1d(20),
            torch.nn.ReLU(),
            torch.nn.Linear(20, num_classes),
        )

    def forward(self, x):
        x = self.cnn_layers(x)
        x = torch.flatten(x, start_dim=1)
        logits = self.fc_layers(x)
        return logits

print("Original 3-block CNN defined.")
````

## Step 4 ŌĆö ­¤ö¦ YOUR TASK: Build a 4-Block CNN

### Background

Look at the shape trace table above. After Block 3's `MaxPool2d`, the spatial dimensions collapse to **1├Ś1** ŌĆö there is no room to add another convolution + pooling on top.

To make space for a 4th convolutional block we need to **remove the `MaxPool2d` from Block 3**. This keeps the feature map at **3├Ś3** after Block 3's convolution, which is just enough for one more `Conv2d(kernel_size=3)` in Block 4.

### What You Need to Change

**Change 1 ŌĆö `cnn_layers`:** In Block 3, remove `torch.nn.MaxPool2d(kernel_size=2)` and add a new Block 4:

```
# Block 4 ŌĆö Input: (32, 3, 3) ŌåÆ Output: (64, 1, 1)
Conv2d(32 ŌåÆ 64, kernel_size=3)
BatchNorm2d(64)
ReLU()
```
> Note: No `MaxPool2d` in Block 4 ŌĆö after Conv2d(k=3) on a 3├Ś3 map the spatial size is already 1├Ś1.

**Change 2 ŌĆö `fc_layers`:** The flattened feature vector is now size **64** (not 32). Update `Linear(32, 20)` ŌåÆ `Linear(64, 20)`.

Everything else ŌĆö `forward()`, the training setup, `LightningModel`, `MnistDataModule` ŌĆö stays exactly the same.

### Shape Trace for the 4-Block CNN

| Layer | Operation | Output Shape (C, H, W) |
|---|---|---|
| Input | ŌĆö | (1, 28, 28) |
| Block 1 | Conv2d(1ŌåÆ3, k=5) + MaxPool | (3, 12, 12) |
| Block 2 | Conv2d(3ŌåÆ16, k=3) + MaxPool | (16, 5, 5) |
| Block 3 | Conv2d(16ŌåÆ32, k=3) ŌĆö **no MaxPool** | (32, 3, 3) |
| Block 4 | Conv2d(32ŌåÆ64, k=3) | (64, 1, 1) |
| Flatten | ŌĆö | (64,) |
| fc_layers | Linear(**64**ŌåÆ20) ŌåÆ BN ŌåÆ ReLU ŌåÆ Linear(20ŌåÆ10) | 10 logits |

````python
# ŌöĆŌöĆ ­¤ö¦ YOUR TASK: Complete the 4-Block CNN below ŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆ
# Follow the two changes described above.
# Only edit the lines marked with  <-- CHANGE THIS

class PyTorchCNN4Block(torch.nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.cnn_layers = torch.nn.Sequential(

            # Block 1 ŌĆö unchanged
            torch.nn.Conv2d(1, 3, kernel_size=5),
            torch.nn.BatchNorm2d(3),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2),

            # Block 2 ŌĆö unchanged
            torch.nn.Conv2d(3, 16, kernel_size=3),
            torch.nn.BatchNorm2d(16),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2),

            # Block 3 ŌĆö remove MaxPool2d here
            torch.nn.Conv2d(16, 32, kernel_size=3),
            torch.nn.BatchNorm2d(32),
            torch.nn.ReLU(),
            # torch.nn.MaxPool2d(kernel_size=2),   # <-- CHANGE THIS (delete this line)

            # Block 4 ŌĆö add your new block here
            torch.nn.Conv2d(32, 64, kernel_size=3), # <-- ADD Conv2d layer   (32 ŌåÆ 64 channels, kernel_size=3)
            torch.nn.BatchNorm2d(64), # <-- ADD BatchNorm2d
            torch.nn.ReLU() # <-- ADD ReLU
        )

        self.fc_layers = torch.nn.Sequential(
            torch.nn.Linear(64, 20),   # <-- CHANGE THIS (update input size)
            torch.nn.BatchNorm1d(20),
            torch.nn.ReLU(),
            torch.nn.Linear(20, num_classes),
        )

    def forward(self, x):
        x = self.cnn_layers(x)
        x = torch.flatten(x, start_dim=1)
        logits = self.fc_layers(x)
        return logits
````

````python
# ŌöĆŌöĆ Quick shape check ŌĆö do not modify ŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆŌöĆ
# Note: we call .eval() so BatchNorm uses running stats instead of batch stats.
# (With a 1├Ś1 spatial output, batch-mode BN needs >1 sample ŌĆö eval() avoids this.)
dummy_input = torch.randn(1, 1, 28, 28)   # 1 grayscale 28├Ś28 image

try:
    test_model = PyTorchCNN4Block(num_classes=10)
    test_model.eval()   # switch to inference mode for the shape check
    with torch.no_grad():
        output = test_model(dummy_input)
    print(f"Output shape: {output.shape}")
    assert output.shape == torch.Size([1, 10]), "Shape mismatch ŌĆö recheck your changes!"
    print("Ō£ģ Shape check passed! Your 4-block CNN is correctly built.")
except Exception as e:
    print(f"ŌØī Error: {e}")
    print("   Re-read the shape trace table and check your Block 3 and fc_layers changes.")
````
